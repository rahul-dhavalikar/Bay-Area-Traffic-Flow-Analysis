## Question 3

```r
mst = minimum.spanning.tree(stockNetwork)
```

```r
plot(degree.distribution(mst), type='l')
plot(degree.distribution(mst))
degreeVector = degree(mst)
histRes = hist(degreeVector, col = 'orange', breaks = c(1:495))
```

```r
mst
```

```r
ecount(mst)
```

```r
vcount(mst)
```

```r
plot(mst, vertex.label=NA, vertex.size = 4, edge.color = 'black')
```

```r
plot(mst,vertex.label=NA,layout=layout.reingold.tilford(mst, circular=T),vertex.size=3,vertex.color=V(mst)$color)
legend("topright", legend=colorNames, col = colorMap   , bty = "o", pch=20 , pt.cex = 1.5, cex = 0.5 , horiz = FALSE)
```

```r
diameter(mst)
```

```r
sectorData = read.csv('finance_data/Name_sector.csv', colClasses = "character")
```

```r
vertices = V(mst)$name
vertexIDs = V(mst)
```

```r
sectors = c(unique(sectorData['Sector']))$Sector
```

```r
sectors
```

```r
colors = rainbow(length(sectors))
```

```r
colors
```

```r
i = 1
for(vertex in vertices){
#     print(vertex)
    x = which(sectorData['Symbol'] == vertex)
    xSector = sectorData[x, 'Sector']
    V(mst)[V(mst)$name == vertex]$sector = xSector
}
```

```r
colorMap = c()
for(s in sectors){
    print(s)
    V(mst)[V(mst)$sector == s]$color = colors[which(sectors == s)]
    print(colors[which(sectors == s)])
    colorMap[[s]] = colors[which(sectors == s)]
}
```

```r
colorMap
```

```r
plot(mst, vertex.label=NA, vertex.size = 4, edge.color = 'black')
legend("topright", legend=colorNames, col = colorMap   , bty = "o", pch=20 , pt.cex = 1.5, cex = 0.5 , horiz = FALSE)
```

```r
colorNames = names(colorMap)
```

## Question 4

```r
adListCustom = c()
for (vertex in vertices){
    n = neighbors(mst, vertex)
    neighborList = c(vertex)
    for (i in 1: length(n)){
        neighborList = c(neighborList, n[[i]]$name)
    }
    strRep = toString(neighborList)
    adListCustom = c(adListCustom, strRep)
}
```

```r
file = "mstAdList_v1.txt"
writeLines(adListCustom, file)
```
